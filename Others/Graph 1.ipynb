```python
import plotly.graph_objects as go
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
fig=go.Figure(data=go.Scatter(x=x,y=y))
fig.show()
```

```python
import plotly.graph_objects as go 
import numpy as np 
  
x = np.outer(np.linspace(-2, 2, 100), np.ones(100)) 
y = x.copy().T 
z = np.cos(x ** 2 + y ** 2) 
  
fig = go.Figure(data=[go.Surface(x=x, y=y, z=z)]) 
  
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
h_water,D=np.mgrid[1:3.3:100j,1:4.65:100j]
x=h_water
y=D
m_water=50.92
mi=m_water*np.tanh(0.866*D/h_water)/(D/h_water*0.866)
z=mi
fig=go.Figure(data=go.Surface(x=x,y=y,z=z))
fig.show()
```
