```python
from sympy import *
B = Symbol('B',imaginary=True)
L1 = 4.35
L2 = 4.65
L = L1+L2+.4
R1 = (B-0.4/2)*.6*.4
R2_1 = (L1-.4)*B*.6
R2_2 = (L2-.4)*B*.6
B1 = 0.4*.6*B
B2_1 = (L1-.4)*.6*B
B2_2 = (L2-.4)*.6*B
P1 = 40+2*R1+B1
P2 = 62+2*(R1+(R2_1+R2_2))+B1+B2_1+B2_2
M_R2 = -R2_1*(L1-.4)/2+R2_2*(L2-.4)/2  # Extra moment due to R2
P3 = 30+2*R1+B1
P = P1+P2+P3
P
```

```python
M1 = 10
M2 = -6
M3 = -8
eL = abs((-P1*L1+M3+P3*L2+M2+M1+M_R2).evalf(2))/P
eL.evalf(2)
```

```python
phi = 34
y = 1.5  # Density of soil in t/m3
Df = 1.2  # Deep foundation

Nq = exp(pi*tan(phi*pi/180))*tan((45+phi/2)*pi/180)**2
Ny = (Nq-1)*tan(1.4*phi*pi/180)
iq = 1
iy = 1
eB=0
B_ef = B-2*eB
L_ef = L-2*eL
sy=1-0.2*B_ef/L_ef
qu = iq*y*Df*Nq+0.5*sy*iy*y*B_ef*Ny
qd = qu/3
simplify(qd.evalf(2))
```

```python
po=P/(L*B)
x=L/2
p=po*(1+3*eL/x)
simplify(p.evalf(2))
```

```python
solve(Eq(simplify(qd.evalf(2)), simplify(p.evalf(2))), B)
```
