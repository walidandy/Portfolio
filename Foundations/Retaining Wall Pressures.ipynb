*EMPUJE ACTIVO (SIN SISMO)*

```python
def kp(beta,alpha,phi,delta):
    return (sin((beta-phi)*pi/180)**2)/(sin(beta*pi/180)**2*sin((beta+delta)*pi/180)*(1-sqrt((sin((phi+delta)*pi/180)*sin((phi+alpha)*pi/180))/(sin((beta+delta)*pi/180)*sin((beta+alpha)*pi/180))))**2)
def Empuje(densidad,H,kp_ka): # Pp (kg/m) , densidad (kg/m3), H (m), kp (adim)
    return 1/2*densidad*H**2*kp_ka
```
